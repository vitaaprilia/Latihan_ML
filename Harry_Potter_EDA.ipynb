```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Harry_Potter_Scraped_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.rename(columns = {"Copies Sold in UK" : "Copies Sold in UK in millions",
                           "Copies Sold WorldWide" : "Copies Sold WorldWide in millions"}, inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
df['Copies Sold in UK in millions']
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,6))
plt.title("Copies Sold in UK")
sns.set(font_scale=1)
sns.barplot(data=df, x="Copies Sold in UK in millions", y="Book Name", hue="Publish Year")
```

```python
df['Copies Sold WorldWide in millions']
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,6))
plt.title("Copies Sold in Worldwide")
sns.set(font_scale=1)
sns.barplot(data=df, x="Copies Sold WorldWide in millions", y="Book Name", hue="Publish Year")
```

```python
df
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,6))
plt.title("Page in a Book")
sns.set(font_scale=1)
sns.barplot(data=df, x="Pages", y="Book Name", hue="Total Awards Won")
```

```python
# converting column "Words" of a DataFrame into Numeric
df['Words'] = df['Words'].str.replace(',', '').astype(float)
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,8))
plt.title("Words in a Book")
sns.set(font_scale=1)
sns.barplot(data=df, y="Book Name", x="Words",hue="Total Awards Won")
```
