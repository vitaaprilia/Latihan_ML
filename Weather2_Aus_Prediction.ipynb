# **Data Cleaning**

```python
# Mengisi null value pada kolom Rainfall, RainToday, dan RainTomorrow. Untuk kolom Rainfall (numerik) diisi dengan rata-rata dan untuk kolom RainToday & RainTomorrow diisi dengan modus.
df['Rainfall'] = df['Rainfall'].fillna(df['Rainfall'].mean())
df['RainToday'] = df['RainToday'].fillna(df['RainToday'].mode()[0])
df['RainTomorrow'] = df['RainTomorrow'].fillna(df['RainTomorrow'].mode()[0])
```
