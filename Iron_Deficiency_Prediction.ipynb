## **4. Mengencoding nilai kategori pada kolom Code menjadi numerik**

```python
# The dataset does not contain data relating the deaths,
#  hence just exploration of inter-feature representations
#  One hot Encoding digunakan untuk mengubah nilai kategori(banyak) menjadi numerik.
unique = pd.get_dummies(df['Code'])
unique
```

```python
df = df.drop(['Entity'], axis=1)
```

## **6. Data Training and Data Testing**

```python
x = df.drop([' Iron deficiency'],axis=1)
y = df[' Iron deficiency']
```

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,
                                                 y,
                                                 test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings("ignore")
kn = KNeighborsRegressor()
kn.fit(x_train, y_train).score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train).score(x_test,y_test)
```

```python
from xgboost import XGBRegressor
import warnings
warnings.filterwarnings("ignore")
xgb = XGBRegressor()
xgb.fit(x_train,y_train).score(x_test,y_test)
```
