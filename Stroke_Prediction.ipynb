```python
dff['ever_married'] = dff['ever_married'].map({
    'Yes' : 0,
    'No' :1
})
```

```python
dff['work_type'] = dff['work_type'].map({
    'Private' : 0,
    'Self-employed' : 1,
    'children' : 2,
    'Govt_job' :3,
    'Never_worked' :4
})
```

```python
dff['Residence_type'].value_counts()
```

```python
dff['Residence_type'] = dff['Residence_type'].map({
    'Urban' : 0,
    'Rural' : 1
})
```

```python
x = dff.drop(['stroke'], axis=1)
y = dff['stroke']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
model = StandardScaler()
x_train= model.fit_transform(x_train)
x_test = model.fit_transform(x_test)
dff
```

```python
dff.info()
```

```python
def models(x_train,y_train):
  # logistic regression
  from sklearn.linear_model import LogisticRegression
  lr = LogisticRegression()
  lr.fit(x_train,y_train)

  # Random Forest 
  from sklearn.ensemble import RandomForestClassifier
  rf = RandomForestClassifier()
  rf.fit(x_train,y_train)

  # XGBClassifier
  from xgboost import XGBClassifier
  xgb = XGBClassifier()
  xgb.fit(x_train,y_train)

  print('lr: ', lr.score(x_train,y_train))
  print('rf: ',rf.score(x_train,y_train))
  print('xgb: ', xgb.score(x_train,y_train))

  return lr,rf,xgb
```

```python
model = models(x_train,y_train)
```

```python
y_pred = model[1].predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
for i in range(len(model)):
  print('Model: ', i)
  print('Accuracy: ', accuracy_score(y_pred, model[i].predict(x_test)))
  print(classification_report(y_pred, model[i].predict(x_test)))
  # print(confusion_matrix(y_pred, model[i].predict(x_test)))
```
